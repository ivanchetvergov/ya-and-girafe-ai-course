```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import datasets, neighbors
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

# Toy task - only 2 features

```python
classification_problem = datasets.make_classification(
    n_samples=100,
    n_features=2,
    n_informative=2,
    n_classes=3,
    n_redundant=0,
    n_clusters_per_class=1,
    random_state=3,
)
```

```python
def plot_dataset(dataset: (np.ndarray, np.ndarray)): # type: ignore
    colors = ListedColormap(["red", "blue", "yellow"])

    plt.figure(figsize=(16, 9))
    plt.grid()
    plt.scatter(dataset[0][:, 0], dataset[0][:, 1], c=dataset[1], cmap=colors, s=100)
    plt.show()
```

```python
plot_dataset(classification_problem)
```

```python
train_data, test_data, train_labels, test_labels = train_test_split(
    *classification_problem,
    test_size=0.3,
    random_state=1,
)
```

# Let's build our model!

```python
clf = neighbors.KNeighborsClassifier()
clf.fit(train_data, train_labels)
```

```python
predictions = clf.predict(test_data)
accuracy_score(test_labels, predictions)
```

### And visualize

```python
def make_meshgrid(
    data: np.ndarray,
    step: float = 0.05,
    border: float = 0.5,
):
    x_min, x_max = data[:, 0].min() - border, data[:, 0].max() + border
    y_min, y_max = data[:, 1].min() - border, data[:, 1].max() + border
    return np.meshgrid(np.arange(x_min, x_max, step), np.arange(y_min, y_max, step))
```

```python
def plot_decision_surface(estimator, train_data, train_labels, test_data, test_labels):
    colors = ListedColormap(["red", "blue", "yellow"])
    light_colors = ListedColormap(["lightcoral", "lightblue", "lightyellow"])

    # fit model
    estimator.fit(train_data, train_labels)

    # set figure size
    fig = plt.figure(figsize=(16, 6))
    fig.suptitle(estimator)

    # plot decision surface on the train data
    plt.subplot(1, 2, 1)
    xx, yy = make_meshgrid(train_data)
    mesh_predictions = np.array(estimator.predict(np.c_[xx.ravel(), yy.ravel()])).reshape(xx.shape)
    plt.pcolormesh(xx, yy, mesh_predictions, cmap=light_colors, shading="auto")
    plt.scatter(train_data[:, 0], train_data[:, 1], c=train_labels, s=100, cmap=colors)
    acc = accuracy_score(train_labels, estimator.predict(train_data))
    plt.title(f"Train data, accuracy={acc:.2f}")

    # plot decision surface on the test data
    plt.subplot(1, 2, 2)
    plt.pcolormesh(xx, yy, mesh_predictions, cmap=light_colors, shading="auto")
    plt.scatter(test_data[:, 0], test_data[:, 1], c=test_labels, s=100, cmap=colors)
    acc = accuracy_score(test_labels, estimator.predict(test_data))
    plt.title(f"Test data, accuracy={acc:.2f}")
```

```python
estimator = neighbors.KNeighborsClassifier(n_neighbors=1)

plot_decision_surface(estimator, train_data, train_labels, test_data, test_labels)
```

## Or not so good...

```python
def train_knn_classifier(dimensions, n_classes):
    scores = []

    for dim in dimensions:
        problem = datasets.make_classification(
            n_samples=1000,
            n_features=dim,
            n_informative=dim // 2,
            n_classes=5,
            n_redundant=dim // 2,
            n_clusters_per_class=1,
            random_state=42,
        )

        train_data, test_data, train_labels, test_labels = train_test_split(
            problem[0],
            problem[1],
            test_size=0.3,
            random_state=1,
        )

        clf = neighbors.KNeighborsClassifier(n_neighbors=5)
        clf.fit(train_data, train_labels)

        predictions = clf.predict(test_data)
        acc = accuracy_score(test_labels, predictions)

        scores.append(acc)

    plt.figure(figsize=(12, 7))
    plt.plot(dimensions, scores)
    plt.grid()
    plt.show()
```

```python
train_knn_classifier([10, 20, 50, 100, 500, 1000], 5)
```

# Bonus area

**Those who gets accuracy of at least 0.8 recieves bonus (0.5 max)**

```python
# YOUR CODE HERE
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

metrics_to_test = ['euclidean', 'manhattan', 'chebyshev']

param_grid = {'n_neighbors': np.arange(1, 21),      #  K  1  20
              'metric' : metrics_to_test,           # 
              'weights' : ['uniform', 'distance']   #   
              } 
knn_gscv = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='accuracy') # 5- -
knn_gscv.fit(X_train, y_train)

print("---  Grid Search ---")
print("  :", knn_gscv.best_params_) 
print("  (accuracy)  -:", knn_gscv.best_score_) 

print(" K:", knn_gscv.best_params_['n_neighbors'])
```

###   

```python
best_clf = knn_gscv.best_estimator_

pred_test = best_clf.predict(X_test)
final_accuracy = accuracy_score(y_test, pred_test)

print(f"test (K=1): {final_accuracy}") 
```

###  ,  . 

```python
from sklearn.metrics import confusion_matrix, classification_report

best_clf = knn_gscv.best_estimator_
y_pred = best_clf.predict(X_test)

conf_matrix = confusion_matrix(y_test, y_pred)
print("--- (Confusion Matrix) ---")
print(conf_matrix)
print(classification_report(y_test, y_pred))
```
